```python
import pandas as pd
import numpy as np
import tensorflow as tf
from keras.layers import LSTM, Embedding, Dense, Bidirectional
data_url = tf.keras.utils.get_file('shakespeare.txt', 'https://storage.googleapis.com/download.tensorflow.org/data/shakespeare.txt')
dataset_text = open(data_url, 'rb').read().decode(encoding='utf-8').lower()
print(dataset_text)
```

```python
characters = sorted(set(dataset_text))
```

```python
characters
```

```python
# Creating a mapping from unique characters to indices
char2idx = dict((c,i) for i,c in enumerate(characters))
char2idx
```

```python
idx2char = dict((i,c) for i,c in enumerate(characters))
idx2char
```

```python
sentences = []
next_char = []

SEQ_LENGTH = 40
STEP_SIZE = 3

for i in range(0, len(dataset_text)-SEQ_LENGTH, STEP_SIZE):
  sentences.append(dataset_text[i:i+SEQ_LENGTH])
  next_char.append(dataset_text[i+SEQ_LENGTH])

x = np.zeros((len(sentences), SEQ_LENGTH,
              len(characters)), dtype=np.bool)
y = np.zeros((len(sentences),
              len(characters)), dtype=np.bool)


for i, sentence in enumerate(sentences):
  for t, char in enumerate(sentence):
    x[i, t, char2idx[char]] = 1
  y[i, char2idx[next_char[i]]] = 1
```

```python
model = tf.keras.Sequential()
model.add(LSTM(128,
               input_shape=(SEQ_LENGTH,
                            len(characters))))
model.add(Dense(len(characters),activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',
              optimizer=tf.optimizers.RMSprop(lr=0.01))

history = model.fit(x,
                    y,
                    batch_size=256,
                    epochs=100)
```

```python
import random 
def sample(preds, temperature=1.0):
    preds = np.asarray(preds).astype('float64')
    preds = np.log(preds) / temperature
    exp_preds = np.exp(preds)
    preds = exp_preds / np.sum(exp_preds)
    probas = np.random.multinomial(1, preds, 1)
    return np.argmax(probas)


def generate_text(length, temperature):
    start_index = random.randint(0, len(dataset_text) - SEQ_LENGTH - 1)
    generated = ''
    sentence = dataset_text[start_index: start_index + SEQ_LENGTH]
    generated += sentence
    for i in range(length):
        x_predictions = np.zeros((1, SEQ_LENGTH, len(characters)))
        for t, char in enumerate(sentence):
            x_predictions[0, t, char2idx[char]] = 1

        predictions = model.predict(x_predictions, verbose=0)[0]
        next_index = sample(predictions,
                                 temperature)
        next_character = idx2char[next_index]

        generated += next_character
        sentence = sentence[1:] + next_character
    return generated
```

```python
print("----------0.2--------")
print(generate_text(300, 0.2))
print("----------0.4--------")
print(generate_text(300, 0.4))
print("----------0.5--------")
print(generate_text(300, 0.5))
print("----------0.6--------")
print(generate_text(300, 0.6))
print("----------0.7--------")
print(generate_text(300, 0.7))
print("----------0.8--------")
print(generate_text(300, 0.8))
```
